```python
%matplotlib widget
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

gme_data = yf.download("GME" , start = '2019-02-07' , end = '2021-02-07',  interval = "1d")
gme_data = gme_data.reset_index()
gme_data.head(10)
```

```python
gme = yf.Ticker("GME")
gme.info
```

```python
gme.major_holders
```

```python
gme.institutional_holders
```

```python
vol_gme = gme_data.plot(x='Date', y="Volume",title="GameStop Stock Volume \nBy Day \nfrom Feb 2019 to Feb 2021");
vol_gme.set_xlabel("Date")
vol_gme.set_ylabel('Volume (1E8)')
plt.show()
```

```python
high_gme = gme_data.plot(x='Date', y="High",title="GameStop Stock High \nBy Day \nfrom Feb 2019 to Feb 2021");
high_gme.set_xlabel("Date")
high_gme.set_ylabel('Highest Price ($)')
plt.show()
```
