```python
%matplotlib widget
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
clean_gme = 'Data/Combined GME Hourly Data.csv'
gme_df = pd.read_csv(clean_gme)
gme_df.head(20)
```

```python
vol_gme = plt.bar(gme_df['time'], gme_df["volume"])
vol_gme.set_xlabel("Date&Time")
vol_gme.set_ylabel('Volume (1E8)')
plt.show()
```
