# Stock Market Graphing

```python
# Dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
from datetime import datetime
```

```python
# Load in csv
GameStop_Stock_df = pd.read_csv("../Data/GME from 2002 to 2021.csv")
```

```python
print(GameStop_Stock_df.head())
GameStop_Stock_df.tail()
```

```python
Line_Chart = GameStop_Stock_df.plot(x="Date", y="Close",title="GameStop Stock Price Close \nBy Day");
Line_Chart.set_ylabel("Price \($)")
plt.show()
```

```python
Line_Chart = GameStop_Stock_df.plot(x="Date", y="Volume",title="GameStop Stock Volume\nBy Day");
Line_Chart.set_ylabel("Stock Trades")
plt.show()
```
